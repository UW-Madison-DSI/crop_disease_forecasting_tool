## Call Wisconet Wrappers

### Python - example

```python
import requests

url = "https://connect.doit.wisc.edu/pywisconet_wrapper/ag_models_wrappers/wisconet"
params = {
    "forecasting_date": "2025-03-20",
    "risk_days": 1
}
headers = {
    "accept": "application/json"
}

response = requests.get(url, params=params, headers=headers)
print(response.status_code)
response.json()
```
